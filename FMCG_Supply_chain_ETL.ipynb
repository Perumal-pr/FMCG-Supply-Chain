```python
# Connecting to the MYSQL server to upload the tables in db
from sqlalchemy import create_engine
import urllib.parse

host = 'localhost'
port = '3306'
username = 'root'
password = urllib.parse.quote_plus("Password@123")
database = 'SupplyChain'


engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}")
```

```python
table_name = "customers"
dim_customers.to_sql(table_name,engine,if_exists="replace",index=False)

print(f" Data Successfully loaded into table name {table_name} in database {database}.")
```

```python
date = 'dates'
dim_date.to_sql(date,engine,if_exists="replace",index=False)

print(f" Data Successfully loaded into table name {date} in database {database}.")
```

```python
products="products"
dim_products.to_sql(products,engine,if_exists="replace",index=False)

print(f" Data Successfully loaded in table {products} in database {database}.")
```

```python
targets_orders="targets_orders"
dim_targets_orders.to_sql(targets_orders,engine,if_exists="replace",index=False)

print(f" Data Successfully loaded in table {targets_orders} in database {database}.")
```

```python
order_lines = "order_lines"

fact_order_lines.to_sql(order_lines,engine,if_exists="replace",index=False)

print(f" Data loaded successfully into table {order_lines} in database {database}.")
```

```python
orders_aggregate = "orders_aggregate"

fact_orders_aggregate.to_sql(orders_aggregate,engine,if_exists="replace",index=False)

print(f" Data loaded successfully into table {orders_aggregate} in database {database}.")
```
